```python
from data.sensors.sensor_loader import SensorLoader

sl = SensorLoader()
sensors_df = sl.load_sensors()
sensors_df.index.name='deviceid'
sensors_df.head(5)
```

```python
# smalldf = sensors_df.iloc[:10,:10]
# sensors_df_pm2_5 = smalldf

sensors_df_pm2_5 = sensors_df
print(sensors_df_pm2_5.head(5))
```

```python
import json


def extract_cell(key:str):
    def extract_cell_with_key(x):
        k = key
        if type(x) == dict:
            if k in x:
                return x[k]
            return 0
        return 0
    return extract_cell_with_key

pm2_5_func = extract_cell('pm2_5')

for column in sensors_df_pm2_5.columns:
    sensors_df_pm2_5[column] = sensors_df_pm2_5.loc[:,column].apply(pm2_5_func)
```

```python
from data.sensors.sensor_loader import SensorLoader

sl = SensorLoader()
sensors_geo_df = sl.load_devices_geo()
sensors_geo_df.index.name='deviceid'
sensors_geo_df.head(5)
```

```python
# join device data and device geo with deviceid

joined_sensors_df = sensors_geo_df.join(sensors_df_pm2_5, on="deviceid")
joined_sensors_df.head(5)
```

```python
joined_sensors_lanlot_df = joined_sensors_df[['lat', 'lon', '2023-06-12T16:01:00.000Z']]
joined_sensors_lanlot_df.head(5)
```

```python
joined_sensors_lanlot_df.rename(columns = {'2023-06-12T16:01:00.000Z':'pm2_5'}, inplace = True)
joined_sensors_lanlot_df.head(5)
```

```python
joined_sensors_lanlot_df_nonna = joined_sensors_lanlot_df.dropna()
```

```python
joined_sensors_lanlot_df_nonna[joined_sensors_lanlot_df_nonna['lat'].isnull() | joined_sensors_lanlot_df_nonna['lon'].isnull() ]
```

```python
#joined_sensors_lanlot_df_nonna
# apply pm2_5 with labels

joined_sensors_lanlot_df_good = joined_sensors_lanlot_df_nonna.loc[joined_sensors_lanlot_df_nonna['pm2_5'] < 25,:]
joined_sensors_lanlot_df_fair = joined_sensors_lanlot_df_nonna.loc[(joined_sensors_lanlot_df_nonna['pm2_5'] >= 25) & (joined_sensors_lanlot_df_nonna['pm2_5'] < 50),:]
joined_sensors_lanlot_df_poor = joined_sensors_lanlot_df_nonna.loc[(joined_sensors_lanlot_df_nonna['pm2_5'] >= 50) & (joined_sensors_lanlot_df_nonna['pm2_5'] < 100),:]
joined_sensors_lanlot_df_very_poor = joined_sensors_lanlot_df_nonna.loc[(joined_sensors_lanlot_df_nonna['pm2_5'] >= 100),:]
```

```python
from bokeh.io import show
from bokeh.plotting import gmap
from bokeh.models import GMapOptions
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook
output_notebook()

bokeh_width, bokeh_height = 1280,720
api_key = ""
def plot(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='air pollution map', 
             width=bokeh_width, height=bokeh_height)
    
    # good_sensor_datasource = ColumnDataSource(joined_sensors_lanlot_df_good)
    # center = p.circle('lon', 'lat', size=1, alpha=0.2, 
    #                   color='green', source=good_sensor_datasource)
      
    center = p.square('lon', 'lat', size=4, alpha=1, 
                      color='green', source=ColumnDataSource(sepd_df))
    center = p.square('lon', 'lat', size=4, alpha=1, 
                      color='green', source=ColumnDataSource(emspd_df))
    center = p.square('lon', 'lat', size=4, alpha=1, 
                      color='green', source=ColumnDataSource(jhspd_df))
    
    
    center = p.circle('lon', 'lat', size=10, alpha=1, 
                      color='yellow', source=ColumnDataSource(joined_sensors_lanlot_df_fair))


    center = p.circle('lon', 'lat', size=4, alpha=1, 
                      color='red', source=ColumnDataSource(joined_sensors_lanlot_df_poor))

    show(p)
    return p

lat, lon = 23.88568, 120.888351
p = plot(lat, lon, zoom=8)
```
